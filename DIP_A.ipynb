```python
# Imports
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import the image
img = cv2.imread('/content/IRIS.jpg') 
img_gray=cv2.imread('/content/IRIS.jpg',0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Plot the image
plt.imshow(img)
plt.show()
print(img.shape)
```

```python

plt.imshow(img_gray,'gray')
plt.show()
```

```python
#binary thresholding
ret, img_binary = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)

# Plot the images
#images = [img, img_binary]
#titles = ['Original image', 'THRESH_BINARY']
plt.imshow(img_binary)
plt.show()
```

```python
#salt and pepper noise
import cv2
import numpy as np
from skimage.util import random_noise
 

# Add salt-and-pepper noise to the image.
noise_img = random_noise(img, mode='s&p',amount=0.3)
 
# The above function returns a floating-point image
# on the range [0, 1], thus we changed it to 'uint8'
# and from [0,255]
noise_img = np.array(255*noise_img, dtype = 'uint8')
 
# Display the noise image
#cv2.imshow('blur',noise_img)
#cv2.waitKey(0)

plt.imshow(noise_img)
plt.show()
```

```python
#salt and pepper noise on gray image
import cv2
import numpy as np
from skimage.util import random_noise
 

# Add salt-and-pepper noise to the image.
noise_img2 = random_noise(img_gray, mode='s&p',amount=0.3)
 
# The above function returns a floating-point image
# on the range [0, 1], thus we changed it to 'uint8'
# and from [0,255]
noise_img2 = np.array(255*noise_img2, dtype = 'uint8')
 
# Display the noise image
#cv2.imshow('blur',noise_img)
#cv2.waitKey(0)

plt.imshow(noise_img2,'gray')
plt.show()
```

```python
#median filter
new_image = cv2.medianBlur(noise_img, 9)
plt.imshow(new_image)
plt.show()
```

```python

from matplotlib import pyplot as plt
from PIL import Image, ImageFilter
%matplotlib inline
image = cv2.cvtColor(noise_img, cv2.COLOR_BGR2HSV) # convert to HSV
figure_size = 9 # the dimension of the x and y axis of the kernal.
new_image = cv2.blur(image,(figure_size, figure_size))
plt.figure(figsize=(11,6))
plt.subplot(122), plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_HSV2RGB)),plt.title('Mean filter')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
def plot_im(img, title):
    plt.figure(figsize=(10,10))
    plt.imshow(img)
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
def mean_kernel_smoothing(img, sz):
    kernel = np.ones((sz, sz)) / (sz ** 2)
    dst = cv2.filter2D(img, -1, kernel)

    plot_im(dst, str(sz) + "X" + str(sz) + " mean kernel")

mean_kernel_smoothing(noise_img,25)
mean_kernel_smoothing(noise_img,30)

mean_kernel_smoothing(noise_img,5)

#plt.imshow(mean_img)
#plt.show()
```
